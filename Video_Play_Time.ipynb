```python
import pandas as pd
```

```python
df = pd.read_csv("play_time_data.csv")
df.head()
```

```python
df1 = df.groupby('category', as_index=False)['Duration_Seconds'].mean()
df1
```

```python
df1.to_csv("avg_play_time_per_category.csv", index=False) 
```

#### What is the average play time of video per category

```python
import matplotlib.pyplot as plt

df1.plot(kind='bar',x='category',y='Duration_Seconds')
```

```python
df2 = df.groupby(['category','Type'], as_index=False)['Duration_Seconds'].mean()
df2
```

#### What is the average play time of video per category per Type (MostViewed, RecentVideos, HighestRated)

```python
df2.to_csv("avg_play_time_per_category_per_type.csv", index=False) 
```

```python
df2.groupby(['category','Type']).sum().plot(kind='bar')
plt.show()
```
